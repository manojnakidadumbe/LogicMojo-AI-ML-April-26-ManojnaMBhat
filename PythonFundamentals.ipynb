Write a program that continuously asks the user to provide an input number.

Print the square of that input number.

The program should stop if the user enters 5.

In case user enters any multiple of 4, skip it.

```python
while True:
  num = int(input("enter no:"))

  if num ==5:
    break
  elif num % 4 == 0:
    continue
  print(num**2)
```

```python
n = 5
for i in range(n):
  for j in range(n-i):
    print(" ", end=' ')
  for k in range(2*i-1):
      print("*", end = ' ')
  print()

for i in range(n-2, -1, -1):
  for j in range(n-i-1):
    print(" ", end = " ")
  for k in range(2*i+1):
    print("*", end = " ")
  print( )
```

# Palindrome

```python
a = "abcd"
b = "madam"

c = a[::-1]
d = b[::-1]

if a ==c:
  print("a is palindrome")

else:
  print("a is not palindrome")

if b ==d:
  print("b is palindrome")

else:
  print("b is not palindrome")
```

```python
text = input("Enter text: ")

reverse = ""

for i in range(len(text)-1, -1, -1):
    reverse += text[i]

if text == reverse:
    print("Palindrome")
else:
    print("Not palindrome")
```
